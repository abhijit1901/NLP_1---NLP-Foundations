# TF-IDF Feature Extraction Pipeline
This notebook demonstrates how to preprocess text data and extract features using the TF-IDF (Term Frequency-Inverse Document Frequency) approach. It covers data loading, cleaning, lemmatization, and vectorization.

## 1. Load SMS Spam Dataset
Read the SMS spam dataset and display its contents.

```python
import pandas as pd
messages=pd.read_csv('SPAMSMSDATASET/SMSSpamCollection',
            sep='\t',names=["label","message"])
```

## 2. Data Cleaning and Preprocessing
Import necessary libraries, download stopwords, and prepare for text cleaning.

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordlemmatize=WordNetLemmatizer()
```

#### Lemmatization and Stopword Removal
Apply lemmatization and remove stopwords from each message to create a cleaned corpus.

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[wordlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

## 4. TF-IDF Vectorization
Convert the cleaned corpus into numerical feature vectors using scikit-learn's `TfidfVectorizer`.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```
